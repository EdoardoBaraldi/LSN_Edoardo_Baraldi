Use the Metropolis algorithm to sample $|\Psi_{1,0,0}(x,y,z)|^2$ and $|\Psi_{2,1,0}(x,y,z)|^2$ **in Cartesian coordinates** using, initially, an uniform transition probability $T(\vec{x}|\vec{y})$. Use the sampled positions to estimate $\left\langle r \right\rangle_{\Psi_{1,0,0}}$ and $\left\langle r \right\rangle_{\Psi_{2,1,0}}$. As usual, use data blocking and give an estimate of the statistical uncertainties.

For example, the ground state wave function is:
$$
\Psi_{1,0,0}(r,\theta,\phi)=
\frac{a_0^{-3/2}}{\sqrt{\pi}} e^{-r/a_0} 
$$
whereas one of the three $2p$ excited state is:
$$
\Psi_{2,1,0}(r,\theta,\phi)=
\frac{a_0^{-5/2}}{8}\sqrt{\frac{2}{\pi}} r e^{-r/2a_0} \cos(\theta)
$$
<span style="color:blue">Expectation values for the radius turns out to be exactly:
$$
\left\langle r \right\rangle_{\Psi_{1,0,0}} = \frac{3}{2}a_0
\quad
\left\langle r \right\rangle_{\Psi_{2,1,0}} = 5 a_0
$$

### <span style="color:blue"> Delta equilibration <span>

```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import pandas as pd
import math

plt.rcParams.update({'font.size': 18})
plt.rcParams["figure.figsize"]= (11,7)

steps_u, a_u, width_u  = np.loadtxt("acceptance_folder/accept_1_0_0_uniform.dat", unpack=True, usecols=(0,1,2), delimiter=",")
steps_g, a_g, width_g  = np.loadtxt("acceptance_folder/accept_1_0_0_gaus.dat", unpack=True, usecols=(0,1,2), delimiter=",")

fig, [ax0, ax1] = plt.subplots(1, 2, figsize=(30 , 12))

ax0.scatter(width_u, a_u, label="Data")
ax1.scatter(width_g, a_g, label="Data")
ax0.plot(width_u, len(a_u)*[a_u[-1]], label="Last a value="+str(a_u[-1]), color='orange')
ax1.plot(width_g, len(a_g)*[a_g[-1]], label="Last a value="+str(a_g[-1]), color='orange')
for i in range(2):
    globals()['ax'+str(i)].set_xlabel("step size", fontsize=20)
    globals()['ax'+str(i)].grid(True)
    globals()['ax'+str(i)].set_ylabel("Acceptance probability", fontsize=20)
    globals()['ax'+str(i)].legend()
ax0.set_title("Acceptance probability for $\Psi_{1,0,0}$ with Uniform distribution",fontsize=25)
ax1.set_title("Acceptance probability for $\Psi_{1,0,0}$ with Gaus distribution", fontsize=25)
```

```python
steps_u, a_u, width_u  = np.loadtxt("acceptance_folder/accept_2_1_0_uniform.dat", unpack=True, usecols=(0,1,2), delimiter=",")
steps_g, a_g, width_g  = np.loadtxt("acceptance_folder/accept_2_1_0_gaus.dat", unpack=True, usecols=(0,1,2), delimiter=",")

fig, [ax0, ax1] = plt.subplots(1, 2, figsize=(30 , 12))

ax0.scatter(width_u, a_u, label="Data")
ax1.scatter(width_g, a_g, label="Data")
ax0.plot(width_u, len(a_u)*[a_u[-1]], label="Last a value="+str(a_u[-1]), color='orange')
ax1.plot(width_g, len(a_g)*[a_g[-1]], label="Last a value="+str(a_g[-1]), color='orange')
for i in range(2):
    globals()['ax'+str(i)].set_xlabel("step size", fontsize=20)
    globals()['ax'+str(i)].grid(True)
    globals()['ax'+str(i)].set_ylabel("Acceptance probability", fontsize=20)
    globals()['ax'+str(i)].legend()
ax0.set_title("Acceptance probability for $\Psi_{2,1,0}$ with Uniform distribution",fontsize=25)
ax1.set_title("Acceptance probability for $\Psi_{2,1,0}$ with Gaus distribution", fontsize=25)
```

I grafici mostrano i vari tentativi nel trovare il corretto step della funzione di probabilit di transizione $T(x'|x)$. L'algoritmo di Metropolis utilizzato per campionare le diverse funzioni di probabilit richiede di definire una probabilit di transizione di prova $T(x'|x)$ da cui ricavare la probabilit di accettazione $A(x'|x)=min \bigl[ 1, \frac{T(x'|x)p(x)}{T(x|x')p(x')}\bigr]$. Affich quest'ultima probabilit rimanga intorno al $50\%$ bisogna calibrare la larghezza dello step con cui si vuole estrarre un nuova posizione $x'$ dalla funzione $T$. Ho scelto quest'ultima in due modi con lo scopo di testare la bont dell'algoritmo di Metropolis: una probabilit di transione uniforme centratra in $x$ con dimensione pari a due volte la largheza dello step ed una probabilit di transizione gaussiana con $\mu=x$ e $\sigma$ pari alla larghezza dello step di transizione.
Per determinare quindi il corretto valore dello step, per ogni ditribuzione da campionare e per ogni $T$ utilizzata, ho inizilizzato lo step stesso ed valutato l'accettanza media di $2\cdot10^4$ estrazioni. Ho valutato l'accettanza facendo un rapporto tra il numero di volte in cui una mossa verso $x'$ fosse accettata e il numero totale di estrazioni di possibili mosse. Svolto questo procedeminento l'ho ripetuto, aumentando di $0.1$ il precendete step testato, fino a quando l'accettanza non venisse compresa tra $0.49$ e $0.51$.

I valori trovati per la dimensione dello step di $T$ sono:
$$\Delta_{\Psi_{1,0,0}}^{u}=2.4 \qquad \Delta_{\Psi_{1,0,0}}^{g}=1.5 \qquad \Delta_{\Psi_{2,1,0}}^{u}=5.9 \qquad \Delta_{\Psi_{2,1,0}}^{g}=3.7 \: .$$

Come si nota gli step nencessari per un'accettanza intorno al $50\%$ per la distribuzione $|\Psi_{2,1,0}(x,y,z)|^2$ sono mediamente pi di quelli per la distribuzione $|\Psi_{1,0,0}(x,y,z)|^2$ poich la prima risulta non sfericamente simmetrica e dotata di due grandi lobi, ad alta probabilit, in direzione Z con una zona a probabilit nulla nell'origine. Questo comportamento fa si che per avere un accettanza ragionevole bisogna muoversi con passi di dimensioni paragonabili alla distanza fra i due lobi.

Da notare come le due distribuzioni di probvabilit siano funzioni nello spazio tridimensionale in coordinate sferiche. Nel mio caso ho sempre utilizzato probabilit test in coordinate cartesiane, ho perci dovuto riscrivere le distribuzioni in queste ultime coordinate.

## $<r>$ for State $\Psi_{1,0,0}$ distribution

```python
r , err_r = np.loadtxt("Data_1_0_0_uniform.dat",  unpack =True, usecols=(2,3), delimiter=",")
r_g , err_r_g = np.loadtxt("Data_1_0_0_gaus.dat",  unpack =True, usecols=(2,3), delimiter=",")


blocks = np.arange(100)


plt.errorbar(blocks+1, r , yerr= err_r, label="unifrom")
plt.errorbar(blocks+1, r_g , yerr= err_r_g, label="gaussian")
plt.axhline(1.5, label="expected", color='r')
plt.xlabel("blocks")
plt.ylabel(r'$<r> [a_0]$')
plt.title(r'$<r>$ with $\Psi_{1,0,0}$ distribution')
plt.legend()
plt.grid(True)
plt.show()
```

Il grafico mostra un confronto del campionamento della distribuzione di probabilit $|\Psi_{1,0,0}(x,y,z)|^2$ utilizzando una probabilit test uniforme ed una gaussiana. Il caloclo del valore medio dalla distanza del centro  stato svolto tramite media a blocchi di $10^6$ estrazioni suddivise in $100$ blocchi. Si nota facilmente che, rispetto al valore analitico aspettato, i due metodi danno rispettivamente una leggera sovrastima e sottostima della distanza dal centro che, tuttavia, tende a ridursi per un elevato numero di blocchi utilizzati nel calcolo della media. Oltre a questo gli errori statistici rendono i valori medi comaptibili con quello ideale gi dopo $80$ blocchi. Le probabilit test implementate equivalgono per quanto riguarda la bont del campionamento.

E' fondamentale sottilineare il punto di partenza del campionamento  in questo l'origine data la simmetria sferica, centra appunto nell'origine della distribuzione di probabilit in question. Eventuali altri punti di partenza da cui dterminare nuove mosse avrebbero semplicemnte richiesto al campionamento pi blocchi per assestarsi intorno al valore ideale.

```python
from mpl_toolkits.mplot3d import axes3d
import matplotlib.pyplot as plt

x_u, y_u, z_u=np.loadtxt("config_1_0_0.final",  unpack =True, usecols=(0,1,2), delimiter=",")
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.scatter(x_u, y_u, z_u, c=z_u, marker='.')

# Set the axis labels
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.set_title("Final configurazion $\Psi_{1,0,0}$ - uniform")

# Rotate the graph
ax.view_init(elev=10, azim=45)  # Set the elevation and azimuth angles

plt.show()
```

Qui in figura viene rappresentata la configurazione finale al centesimo blocco della distribuzione campionata con probabilit test uniforme.

Come nel caso precedente, rispetto al valore analitico aspettato, i due metodi danno rispettivamente una leggera sovrastima e sottostima della distanza dal centro che, tuttavia, tende a ridursi per un elevato numero di blocchi utilizzati.
