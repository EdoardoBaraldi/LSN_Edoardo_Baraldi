### Exercise 02.1
- Compute the following 1D integral via Monte Carlo 
$$I = \int_0^1 \frac{\pi}{2}\cos(\pi x/2) dx = 1$$

1. sampling a uniform distribution in $[0,1]$
2. using importance sampling (i.e. sampling a non-uniform probability in $[0,1]$)

<span style="color:red">Show a picture of both your estimations of $I$</span> and their uncertainties with a large number of *throws* $M$ (e.g. $M\ge 10^4$) as a function of the number of blocks, $N$</font>.

```python
import numpy as np
import matplotlib.pyplot as plt 

media_uni, varianza_uni =np.loadtxt("Data1_1", unpack =True, usecols=(0,1), delimiter=",")
media_corr, varianza_corr =np.loadtxt("Data1_2", unpack =True, usecols=(0,1), delimiter=",")
media_wrong, varianza_wrong =np.loadtxt("Data1_3", unpack =True, usecols=(0,1), delimiter=",")

M=1000000   #number of total throws
N=100       #nuber of blocks
L=int(M/N)  #number of throws in each block

x=np.arange(N)

fig, [ax0, ax1, ax2] = plt.subplots(1, 3, figsize=(27 , 8))

ax0.errorbar(x+1, media_uni, yerr=varianza_uni)
ax0.axhline(y=1., linestyle='dotted', label='true mean', color='r')
ax0.set_title(r'Integral of $\frac{\pi}{2}cos(\frac{\pi x}{2})$ - uniform')
ax0.set_ylabel(r'$\langle f \rangle - 1$', fontsize=16)
ax0.legend()
ax0.grid(True)

ax1.errorbar(x+1, media_corr, yerr=varianza_corr, color='orange')
ax1.axhline(y=1., linestyle='dotted', label='true mean', color='r')
ax1.set_title(r'Integral of $\frac{\pi}{2}cos(\frac{\pi x}{2})$ - $p(x)=2(1-x)$')
ax1.legend()
ax1.grid(True)

ax2.errorbar(x+1, media_wrong, yerr=varianza_wrong, color='green')
ax2.axhline(y=1., linestyle='dotted', label='true mean', color='r')
ax2.set_title(r'Integral of $\frac{\pi}{2}cos(\frac{\pi x}{2})$ - $p(x)=3x^2$')
ax2.legend()
ax2.grid(True)


for i in range(3):
    globals()['ax'+str(i)].set_xlabel("blocks", fontsize=16)
    globals()['ax'+str(i)].grid(True)
    
plt.show()
```

I tre grafici mostrano i risultati della media a blocchi con $10^6$ estrazioni e $100$ blocchi per il calcolo dell'integrale $I = \int_0^1 \frac{\pi}{2}\cos(\pi x/2) dx = 1$. 

Il primo calcolo è stato eseguito campionando una distribuzione di probabilità uniforme da 0 a 1. Esso mostra come il valore medio dei blocchi, per un numerosi blocchi, sia compatibile con il valore reale. 

Il secondo integrale è stato eseguito campionando invece una distribuzione normalizzata $p(x)=2(1-x)$ più simile della distribuzione uniforme all'integranda. E' necessario notare che il valore medio per numerosi blocchi è compatibile con quello reale a meno di 2 errori stastici di distanza.

Il terzo integrale è invece chiaramente il peggiore dei tre a cuasa del campionamento con una una funzione di distribuzione dissimile, per comportamento, all'integranda. Il valor medio per numerosi blocchi comunque è compatibile con quello reale a meno di 3 errori di distanza.

```python
plt.plot(x+1, varianza_uni, label="uniform")
plt.plot(x+1, varianza_corr, label="$p(x)=2(1-x)$")
plt.plot(x+1, varianza_wrong, label="$p(x)=3x^2$")
plt.title(r'$\sigma$ in function of number of blocks')
plt.ylabel(r'$\sigma$', fontsize=16)
plt.xlabel('blocks', fontsize=16)
plt.legend()
plt.grid(True)
```

Gli errori statistici in funzione del numero di blocchi utilizzati al variare del funzione di distribuzione di campionamento dell'integranda mostra come la funzione $p(x)=2(1-x)$ abbia effettivamente abbassato la deviazione standard rispetto ad una funzione di campionamento uniforme, scopo infatti dell'importance sampling. La funzione quadratica usata per il campionamento mostra ancora chiaramente quanto sia inefficiente per l'integranda in questione.

### Exercise 02.2
- <span style="color:red">3D Random Walks (RW) on a cubic lattice and in the continuum:</span> Repeat many times (e.g. say $10^4$) the simulation of a random walk in 3D always starting at the origin

1. on a cubic lattice with lattice constant $a=1$; at each discrete time the walker makes a forward or backward step of length equal to $a$ in one of the 3 principal directions of the lattice: $x$, $y$ or $z$
2. in the continuum; at each discrete time the walker makes a step of length equal to $a(=1)$ along a **random direction** obtained by sampling **uniformly** the solid angle: $\theta \in [0,\pi]$ and $\phi \in [0,2\pi]$

<span style="color:red">Show a picture of $\sqrt{\langle |\vec{r}_N|^2 \rangle_{RW}}$ for both RWs</span>, with their statistical uncertainties, as function of the step $i\in [0,10^2]$.

Suggestion: divide your $M$ throws into a reasonable number of blocks, compute $\sqrt{\langle |\vec{r}_N|^2 \rangle_{RW}}$ as function of the step $i\in [0,10^2]$ in each block and use these values (for fixed $i$) to obtain the average value and its statistical uncertainty.

```python
from scipy.optimize import curve_fit
def sqrN(x,k):
    return k*np.sqrt(x)
```

```python
r_rw, err_r_rw =np.loadtxt("Data2_1", unpack =True, usecols=(0,1), delimiter=",")
r_rw_cont, err_r_rw_cont =np.loadtxt("Data2_2", unpack =True, usecols=(0,1), delimiter=",")

fig, [ax0, ax1] = plt.subplots(1,2, figsize=(20,5))

M=1000000   #number of total throws
N=100       #nuber of blocks
L=int(M/N)  #number of throws in each block
n_step=100  #number of steps

x=np.arange(n_step+1)

popt_lattice, pcov_lattice = curve_fit(sqrN, xdata=x, ydata=r_rw)
popt_continous, pcov_continous = curve_fit(sqrN, xdata=x, ydata=r_rw_cont)

ax0.errorbar(x, r_rw, yerr=err_r_rw)
ax0.plot(x, sqrN(x,popt_lattice), label="$f(N)=k\sqrt{N}$", color="r")
ax1.errorbar(x, r_rw_cont, yerr=err_r_rw_cont)
ax1.plot(x, sqrN(x,popt_continous), label="$f(N)=k\sqrt{N}$", color="r")

for i in range(2):
    globals()['ax'+str(i)].set_ylabel(r'$\sqrt{\langle |\vec{r}_N|^2 \rangle_{RW}}$', fontsize=15)
    globals()['ax'+str(i)].set_xlabel("steps", fontsize=15)
    globals()['ax'+str(i)].grid(True)

ax0.set_title(r'$\sqrt{\langle |\vec{r}_N|^2 \rangle_{RW}}$ in lattice with a=1')
ax1.set_title(r'$\sqrt{\langle |\vec{r}_N|^2 \rangle_{RW}}$ continous with $\rho$=1')
plt.show()

print("  Optimized k (cubic lattice): ",popt_lattice, "\t\t\t","k parameter (continuum): ", popt_continous)
print("  k uncertainty (cubic lattice): ", np.sqrt(np.diagonal(pcov_lattice)), "\t\t","k uncertainty (continuum): ", np.sqrt(np.diagonal(pcov_continous))) 
```

Entrambi i grafici della distanza dall'origine di un Random Walk (RW) in funzione degli steps seguono l'andamento $\sqrt{N}$; infatti i fit fatti con una funzione $f(N)=k\sqrt{N}$ mostrano che per entrambe le situazioni $k\simeq1$. 

Questo fa concludere che i punti in un random walk seguono un andamento descritto da un'equazione di diffusione $$\frac{\partial p}{\partial t} = D\frac{\partial^{2} p}{\partial x^2}$$ dove $$D=\frac{a^2}{2\tau}$$. 

In entrambi i casi proposti $a=1$ dove nel reticolo cubico corrsiponde al lato del reticolo stesso e nel caso continuo corrisponde alla lunghezza del passo $\rho$ in direzione arbitraria. 

Se risolviamo l'equazione differenziale di diffusione imponendo la condizione iniziale $x_0=0$ (ossia partiamo dall'origine) si ottiene una gaussiana centrata nell'origine con $\sigma = \sqrt{N}$ ($N$ numero di steps), che è uguale all'andamento ottenuto dalle simulazioni.
